```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
```

```python
df = pd.read_excel("/Users/GONZALJI/Downloads/motivation.xlsx")
```

```python
df["Possible transactions"] = df["Number of firms"] * (df["Number of firms"] - 1)
```

```python
fig, ax = plt.subplots(1, figsize=(10, 7.5))
sns.lineplot(data=df, x="Year", y="Number of transactions", ax=ax, label="Actual")
sns.lineplot(data=df, x="Year", y="Possible transactions", ax=ax, label="Possible")
ax.set_xticks(range(2008, 2020))
ax.set_ylim([1.e+5, 1.e+10])
ax.set_yscale("log")
plt.savefig("motivation.eps")
```
